```python
data['Hours_Played'] = data.apply(lambda row: row['SessionsPerWeek']*row['AvgSessionDurationMinutes'] , axis=1)
```

# Distribuition 

```python
# Distribution of vehicles based on year of registration
fig, ax = plt.subplots(figsize=(8,6))
sns.distplot(data['Hours_Played'], color="#33cc33",kde=True, ax=ax)
ax.set_title('Distribution of Hours Played', fontsize= 15)
plt.ylabel("Density (KDE)", fontsize= 15)
plt.xlabel("Hours Played", fontsize= 15)
plt.show()
```

# Bar Plots

```python
sns.set_style("white")
g = sns.catplot(x="Location", data=data, kind="count",
                   palette="BuPu", aspect=1.5)
g.ax.xaxis.set_label_text("Location",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Total No of players", fontdict= {'size':16})
g.ax.set_title("Total count of Players per Location",fontdict= {'size':18})
# to get the counts on the top heads of the bar
for p in g.ax.patches:
    g.ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+500))
```
